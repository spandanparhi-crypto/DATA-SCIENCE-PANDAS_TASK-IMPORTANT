```python
print("Before:", df['Gender'].unique())

def clean_gender(x):
    if pd.isna(x):
        return 'Unknown'
    x = str(x).strip().lower()
    if x in ['m', 'male']:
        return 'Male'
    if x in ['f', 'female']:
        return 'Female'
    return 'Unknown'

df['Gender'] = df['Gender'].apply(clean_gender)
print("After:", df['Gender'].unique())
df['Gender'].value_counts()
```

```python
print("Before:", df['City'].unique())

def clean_city(x):
    if pd.isna(x):
        return x
    x = str(x).strip().lower()
    mapping = {
        'hyd': 'Hyderabad', 'hyderabad': 'Hyderabad', 'hyderaba': 'Hyderabad',
        'bangalore': 'Bengaluru', 'bengaluru': 'Bengaluru',
        'chennai': 'Chennai', 'delhi': 'Delhi', 'mumbai': 'Mumbai', 'pune': 'Pune',
        'not available': 'Not Available'
    }
    return mapping.get(x, x.title())

df['City'] = df['City'].apply(clean_city)
print("After:", df['City'].unique())
df['City'].value_counts()
```

```python
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = series[(series < lower) | (series > upper)]
    return outliers, lower, upper

for col in ['Salary', 'Performance_Score']:
    outliers, lower, upper = detect_outliers_iqr(df[col].dropna())
    print(f"{col}: lower bound={lower:.2f}, upper bound={upper:.2f}")
    print(f"  Outlier values: {outliers.tolist()}\n")
```

```python
for col in ['Salary', 'Performance_Score']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower, upper=upper)

df[['Salary', 'Performance_Score']].describe()
```

```python
def clean_employee_data(path):
    df = pd.read_excel(path)

    # Fix shifted header
    df = df.drop(columns=['Age']).rename(columns={'Unnamed: 3': 'Age'})
    df = df.rename(columns={
        'Customer': 'Customer_ID', 'Experienc': 'Experience',
        'Departme': 'Department', 'Joining_D': 'Joining_Date',
        'Performa': 'Performance_Score'
    })
    df = df[['Customer_ID', 'Name', 'Age', 'Gender', 'City', 'Salary',
              'Experience', 'Department', 'Joining_Date', 'Performance_Score']]

    # Handle missing values
    df['Name'] = df['Name'].fillna('Unknown')
    df['City'] = df['City'].fillna('Not Available')
    df['Department'] = df['Department'].fillna('Not Assigned')

    # Remove duplicates
    df = df.drop_duplicates(subset='Customer_ID', keep='first')

    # Clean strings
    df['Name'] = df['Name'].astype(str).str.strip()

    def clean_gender(x):
        if pd.isna(x): return 'Unknown'
        x = str(x).strip().lower()
        if x in ['m', 'male']: return 'Male'
        if x in ['f', 'female']: return 'Female'
        return 'Unknown'
    df['Gender'] = df['Gender'].apply(clean_gender)

    def clean_city(x):
        if pd.isna(x): return x
        x = str(x).strip().lower()
        mapping = {'hyd': 'Hyderabad', 'hyderabad': 'Hyderabad', 'hyderaba': 'Hyderabad',
                   'bangalore': 'Bengaluru', 'bengaluru': 'Bengaluru', 'chennai': 'Chennai',
                   'delhi': 'Delhi', 'mumbai': 'Mumbai', 'pune': 'Pune',
                   'not available': 'Not Available'}
        return mapping.get(x, x.title())
    df['City'] = df['City'].apply(clean_city)

    def clean_dept(x):
        if pd.isna(x): return x
        x = str(x).strip()
        mapping = {'it': 'IT', 'hr': 'HR', 'finance': 'Finance',
                   'sales': 'Sales', 'marketing': 'Marketing',
                   'not assigned': 'Not Assigned'}
        return mapping.get(x.lower(), x.title())
    df['Department'] = df['Department'].apply(clean_dept)

    # Clean Salary
    df['Salary'] = (df['Salary'].astype(str)
                     .str.replace('₹', '', regex=False)
                     .str.replace(',', '', regex=False))

    # Convert datatypes
    for col in ['Age', 'Salary', 'Experience', 'Performance_Score']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Fix invalid values
    df.loc[(df['Age'] < 18) | (df['Age'] > 100), 'Age'] = np.nan
    df.loc[df['Experience'] < 0, 'Experience'] = 0

    # Handle outliers (cap using IQR)
    for col in ['Salary', 'Performance_Score']:
        Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=lower, upper=upper)

    # Dates + feature engineering
    df['Joining_Date'] = pd.to_datetime(df['Joining_Date'], errors='coerce')
    df['Joining_Year'] = df['Joining_Date'].dt.year
    df['Joining_Month'] = df['Joining_Date'].dt.month
    df['Joining_Day'] = df['Joining_Date'].dt.day

    return df

df_clean = clean_employee_data("D:\Data science\Datasets\Pandas_Practice_Dataset.xlsx")
df_clean.head(10)
```

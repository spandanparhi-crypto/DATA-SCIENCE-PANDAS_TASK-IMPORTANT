```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 140)
```

```python
path = "D:\Data science\Datasets\HR_Employee_Dataset.csv.tsv"
df = pd.read_csv(path, sep='\t')                                 
```

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
df['Joining_Year'] = df['Joining_Date'].dt.year
```

```python
print("Dataset shape:", df.shape)
print("=" * 80)
```

# SECTION 1 — BASIC (Q1-10)

```python
print("\nQ1. Employees per Department")
print("-" * 50)
q1 = df['Department'].value_counts()
print(q1)
```

```python
print("\nQ2. Gender distribution (%)")
print("-" * 50)
q2 = (df['Gender'].value_counts(normalize=True) * 100).round(2)
print(q2)
```

```python
print("\nQ3. Average salary of the company")
print("-" * 50)
q3 = df['Salary'].mean()
print(f"Average Salary: {q3:,.2f}")
```

```python
print("\nQ4. Employees older than 50 years")
print("-" * 50)
q4 = df[df['Age'] > 50]
print(f"Count: {len(q4)}")
print(q4[['Employee_ID', 'Name', 'Age', 'Department']].head(10))
```

```python
print("\nQ5. Employees with Performance_Rating = 5")
print("-" * 50)
q5 = df[df['Performance_Rating'] == 5]
print(f"Count: {len(q5)}")
print(q5[['Employee_ID', 'Name', 'Department', 'Performance_Rating']].head(10))
```

```python
print("\nQ6. Employees who joined after 2020")
print("-" * 50)
q6 = (df['Joining_Year'] > 2020).sum()
print(f"Count: {q6}")
```

```python
print("\nQ7. Youngest and oldest employee age")
print("-" * 50)
print(f"Youngest: {df['Age'].min()}  |  Oldest: {df['Age'].max()}")
```

```python
print("\nQ8. Top 10 highest-paid employees")
print("-" * 50)
q8 = df.nlargest(10, 'Salary')[['Employee_ID', 'Name', 'Department', 'Designation', 'Salary']]
print(q8.to_string(index=False))
```

```python
print("\nQ9. Employees per Designation")
print("-" * 50)
q9 = df['Designation'].value_counts()
print(q9)
```

```python
print("\nQ10. Average experience in IT department")
print("-" * 50)
q10 = df[df['Department'] == 'IT']['Experience_Years'].mean()
print(f"Average Experience (IT): {q10:.2f} years")
```

# SECTION 2 — INTERMEDIATE (Q1-10)

```python
print("\nQ1. Department-wise average salary")
print("-" * 50)
i1 = df.groupby('Department')['Salary'].mean().round(2).sort_values(ascending=False)
print(i1)
```

```python
print("\nQ2. Department with highest average performance rating")
print("-" * 50)
i2 = df.groupby('Department')['Performance_Rating'].mean().round(3).sort_values(ascending=False)
print(i2)
print(f"\nHighest: {i2.idxmax()} ({i2.max():.3f})")
```

```python
print("\nQ3. Employees with salary above company average")
print("-" * 50)
company_avg_salary = df['Salary'].mean()
i3 = df[df['Salary'] > company_avg_salary]
print(f"Count: {len(i3)} out of {len(df)} (Company avg = {company_avg_salary:,.2f})")
print(i3[['Employee_ID', 'Name', 'Salary']].head(10))
```

```python
print("\nQ4. Age Group column + counts")
print("-" * 50)
def age_group(age):
    if 20 <= age <= 30:
        return 'Young'
    elif 31 <= age <= 45:
        return 'Mid'
    else:  # 46+
        return 'Senior'

df['Age_Group'] = df['Age'].apply(age_group)
i4 = df['Age_Group'].value_counts()
print(i4)
```

```python
print("\nQ5. Employees with experience > their department's average experience")
print("-" * 50)
dept_avg_exp = df.groupby('Department')['Experience_Years'].transform('mean')
i5 = df[df['Experience_Years'] > dept_avg_exp]
print(f"Count: {len(i5)}")
print(i5[['Employee_ID', 'Name', 'Department', 'Experience_Years']].head(10))
```

```python
print("\nQ6. Top 10 employees by Salary/Experience ratio")
print("-" * 50)
df['Salary_Growth_Ratio'] = (df['Salary'] / df['Experience_Years']).round(2)
i6 = df.nlargest(10, 'Salary_Growth_Ratio')[
    ['Employee_ID', 'Name', 'Department', 'Salary', 'Experience_Years', 'Salary_Growth_Ratio']]
print(i6.to_string(index=False))
```

```python
print("\nQ7. Designation with highest average salary")
print("-" * 50)
i7 = df.groupby('Designation')['Salary'].mean().round(2).sort_values(ascending=False)
print(i7)
print(f"\nHighest: {i7.idxmax()} ({i7.max():,.2f})")
```

```python
print("\nQ8. Departments where average salary > 80,000")
print("-" * 50)
dept_avg_sal = df.groupby('Department')['Salary'].mean().round(2)
i8 = dept_avg_sal[dept_avg_sal > 80000]
print(i8 if len(i8) > 0 else "No department has an average salary above 80,000.")
print("\n(For reference, all department averages:)")
print(dept_avg_sal.sort_values(ascending=False))
```

```python
print("\nQ9. Total salary cost per department")
print("-" * 50)
i9 = df.groupby('Department')['Salary'].sum().sort_values(ascending=False)
print(i9)
```

```python
print("\nQ10. Employees who joined in the same year but different departments")
print("-" * 50)
# Logic: group by Joining_Year, keep years where >1 distinct department is represented
year_dept = df.groupby('Joining_Year')['Department'].nunique()
multi_dept_years = year_dept[year_dept > 1].index
i10 = df[df['Joining_Year'].isin(multi_dept_years)].sort_values(['Joining_Year', 'Department'])
print(f"Years with employees joining across multiple departments: {list(multi_dept_years)}")
print(i10[['Employee_ID', 'Name', 'Joining_Year', 'Department']].head(15).to_string(index=False))
print(f"... total matching rows: {len(i10)}")
```

# SECTION 3 — ADVANCED (Q1-10)

```python
print("\nQ1. Salary above dept average BUT performance below dept average")
print("-" * 50)
dept_avg_sal_t = df.groupby('Department')['Salary'].transform('mean')
dept_avg_perf_t = df.groupby('Department')['Performance_Rating'].transform('mean')
a1 = df[(df['Salary'] > dept_avg_sal_t) & (df['Performance_Rating'] < dept_avg_perf_t)]
print(f"Count: {len(a1)}")
print(a1[['Employee_ID', 'Name', 'Department', 'Salary', 'Performance_Rating']].head(10))
```

```python
print("\nQ2. Salary outliers per department (IQR method)")
print("-" * 50)
outlier_rows = []
for dept, g in df.groupby('Department'):
    Q1, Q3 = g['Salary'].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = g[(g['Salary'] < lower) | (g['Salary'] > upper)]
    print(f"{dept}: Q1={Q1:,.0f}, Q3={Q3:,.0f}, IQR bounds=[{lower:,.0f}, {upper:,.0f}], outliers={len(outliers)}")
    outlier_rows.append(outliers)
a2 = pd.concat(outlier_rows)
print(f"\nTotal outliers across all departments: {len(a2)}")
print(a2[['Employee_ID', 'Name', 'Department', 'Salary']].head(10))
```

```python
print("\nQ3. Promotion risk flag")
print("-" * 50)
dept_median_sal = df.groupby('Department')['Salary'].transform('median')
df['Promotion_Risk'] = (
    (df['Experience_Years'] > 8) &
    (df['Performance_Rating'] >= 4) &
    (df['Salary'] < dept_median_sal)
)
a3 = df[df['Promotion_Risk']]
print(f"Count: {len(a3)}")
print(a3[['Employee_ID', 'Name', 'Department', 'Experience_Years', 'Performance_Rating', 'Salary']].head(10))
```

```python
print("\nQ4. Departments with average experience higher than company-wide average")
print("-" * 50)
company_avg_exp = df['Experience_Years'].mean()
dept_avg_exp_g = df.groupby('Department')['Experience_Years'].mean().round(2)
a4 = dept_avg_exp_g[dept_avg_exp_g > company_avg_exp]
print(f"Company-wide average experience: {company_avg_exp:.2f} years\n")
print(a4.sort_values(ascending=False))
```

```python
print("\nQ5. Correlation: Age, Salary, Experience, Performance_Rating")
print("-" * 50)
a5 = df[['Age', 'Salary', 'Experience_Years', 'Performance_Rating']].corr().round(3)
print(a5)
```

```python
print("\nQ6. High-risk attrition candidates (Age<30, Salary<median, Perf<=2)")
print("-" * 50)
company_median_sal = df['Salary'].median()
a6 = df[(df['Age'] < 30) & (df['Salary'] < company_median_sal) & (df['Performance_Rating'] <= 2)]
print(f"Company median salary: {company_median_sal:,.2f}")
print(f"Count: {len(a6)}")
print(a6[['Employee_ID', 'Name', 'Age', 'Salary', 'Performance_Rating']].head(10))
```

```python
print("\nQ7. Employee ranking within department by salary")
print("-" * 50)
df['Dept_Salary_Rank'] = df.groupby('Department')['Salary'].rank(ascending=False, method='min').astype(int)
a7 = df.sort_values(['Department', 'Dept_Salary_Rank'])
print(a7[['Employee_ID', 'Name', 'Department', 'Salary', 'Dept_Salary_Rank']].groupby('Department').head(3).to_string(index=False))
```

```python
print("\nQ8. Salary inequality (max - min) per department")
print("-" * 50)
a8 = df.groupby('Department')['Salary'].agg(Min='min', Max='max')
a8['Range'] = a8['Max'] - a8['Min']
print(a8.sort_values('Range', ascending=False))
```

```python
print("\nQ9. Cohort analysis — average salary by joining year")
print("-" * 50)
a9 = df.groupby('Joining_Year')['Salary'].agg(Avg_Salary='mean', Count='count').round(2)
print(a9.sort_index())
```

```python
print("\nQ10. Performance segments — avg salary & experience comparison")
print("-" * 50)
def perf_segment(r):
    if r <= 2:
        return 'Low'
    elif r <= 4:
        return 'Medium'
    else:
        return 'High'

df['Performance_Segment'] = df['Performance_Rating'].apply(perf_segment)
a10 = df.groupby('Performance_Segment').agg(
    Avg_Salary=('Salary', 'mean'),
    Avg_Experience=('Experience_Years', 'mean'),
    Count=('Employee_ID', 'count')
).round(2).reindex(['Low', 'Medium', 'High'])
print(a10)
```

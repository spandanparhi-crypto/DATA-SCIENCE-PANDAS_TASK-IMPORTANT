```python
import pandas as pd

# Load Dataset
df = pd.read_csv(r"D:\Data science\Datasets\customer_churn.csv")

print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# Replace Male with M and Female with F

df["gender"] = df["gender"].replace({
    "Male": "M",
    "Female": "F"
})

print(df["gender"].head())
```

```python
sorted_age = df.sort_values("age")

print(sorted_age[["age"]].head())
```

```python
sorted_income = df.sort_values("income", ascending=False)

print(sorted_income[["income"]].head())
```

```python
sorted_index = df.sort_index()

print(sorted_index.head())
```

```python
group_region = df.groupby("region")["income"].mean()

print(group_region)
```

```python
gender_count = df.groupby("gender")["age"].count()

print(gender_count)
```

```python
df["region_avg_income"] = df.groupby("region")["income"].transform("mean")

df["income_diff"] = df["income"] - df["region_avg_income"]

print(df[["region","income","region_avg_income","income_diff"]].head())
```

```python
high_income = df["income"].where(df["income"] > 50000, "Low Income")

print(high_income.head(10))
```

```python
masked_income = df["income"].mask(df["income"] > 50000, "High Income")

print(masked_income.head(10))
```

```python
import pandas as pd

df = pd.read_csv(r"D:\Data science\Datasets\customer_churn.csv")

# 23 Replace
df["gender"] = df["gender"].replace({"Male":"M","Female":"F"})

# 24 Sort Values
print(df.sort_values("age").head())

# 25 Sort Index
print(df.sort_index().head())

# 26 GroupBy
print(df.groupby("region")["income"].mean())

# 27 Transform
df["region_avg_income"] = df.groupby("region")["income"].transform("mean")

# 28 Where
df["income_where"] = df["income"].where(df["income"] > 50000, "Low Income")

# 29 Mask
df["income_mask"] = df["income"].mask(df["income"] > 50000, "High Income")

print(df.head())
```

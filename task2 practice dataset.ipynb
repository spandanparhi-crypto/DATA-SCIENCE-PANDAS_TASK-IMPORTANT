```python
import pandas as pd

# Load Dataset
df = pd.read_excel(r"D:\Data science\Datasets\Data_Train.xlsx")

print("Original Shape :", df.shape)
```

```python
df_airline = df.drop_duplicates(subset=["Airline"])

print(df_airline[["Airline"]])
print("Shape After Removing Duplicate Airlines :", df_airline.shape)
```

```python
df_rename = df.rename(columns={
    "Airline":"Flight_Airline",
    "Source":"From_City",
    "Destination":"To_City"
})

print(df_rename.columns)
```

```python
print("Before :", df["Price"].dtype)

df["Price"] = df["Price"].astype(float)

print("After :", df["Price"].dtype)
```

```python
df["Price"] = pd.to_numeric(df["Price"])

print(df["Price"].dtype)
```

```python
df["Price_Str"] = df["Price"].astype(str)

df["Price_Str"] = pd.to_numeric(df["Price_Str"])

print(df["Price_Str"].dtype)
```

```python
df["Route"] = df["Route"].fillna("Unknown Route")
df["Total_Stops"] = df["Total_Stops"].fillna("0 stop")

print(df.isnull().sum())
```

```python
df.fillna(0, inplace=True)

print(df.isnull().sum())
```
